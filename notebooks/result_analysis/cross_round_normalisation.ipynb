```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
loc = 1 # mean
scale = 2 # std
normal_samples = np.random.normal(loc = loc, scale = scale, size = 1000)
```

```python
plt.hist(normal_samples)
```

# Normalisation using sample mean and std

```python
sample_mean = np.mean(normal_samples)
sample_std = np.std(normal_samples)
normalised_samples_true = (normal_samples - sample_mean)/sample_std
```

```python
plt.hist(normalised_samples_true)
```

```python
normalised_samples_true.mean()
```

```python
normalised_samples_true.std()
```

# Normalisation using random samples

```python
random_samples = np.random.choice(normal_samples, size = 4, replace=False)
```

```python
sample_mean = np.mean(random_samples)
sample_std = np.std(random_samples)
normalised_samples_random = (normal_samples - sample_mean)/sample_std
```

```python
plt.hist(normalised_samples_random)
```

```python
normalised_samples_random.mean()
```

```python
normalised_samples_random.std()
```

## repeat several times

```python
random_mean_list = []
random_std_list = []
repeat_times = 10
for i in range(repeat_times):
    random_samples = np.random.choice(normal_samples, size = 4, replace=False)
    sample_mean = np.mean(random_samples)
    sample_std = np.std(random_samples)
    normalised_samples_random = (normal_samples - sample_mean)/sample_std
    random_mean_list.append(normalised_samples_random.mean())
    random_std_list.append(normalised_samples_random.std())
```

```python
plt.scatter(range(repeat_times),random_mean_list)
plt.hlines(np.mean(random_mean_list), 0, repeat_times)
plt.title('Normalised mean using random samples')
plt.ylabel('normalised mean')
```

```python
plt.scatter(range(repeat_times),random_std_list)
plt.hlines( np.mean(random_std_list), 0, repeat_times)
plt.title('Normalised std using random samples')
plt.ylabel('normalised std')
```

# Normalisation using sample quartiles

```python

repeat_times = 10
quartiles = [0, 25, 50, 75, 100]

sample_quartiles = np.percentile(normal_samples, quartiles)
sample_mean = np.mean(sample_quartiles)
sample_std = np.std(sample_quartiles)
normalised_samples_quartiles = (normal_samples - sample_mean)/sample_std
```

```python
plt.hist(normalised_samples_quartiles)
```

```python
np.mean(normalised_samples_quartiles)
```

```python
np.std(normalised_samples_quartiles)
```
